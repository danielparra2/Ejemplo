```python
#libreria de matematicas
import numpy as np
#libreria para graficas # modulo pyplot
import matplotlib.pyplot  as plt 
#Biblioteca de tabla ,sirve para excel
import pandas as pd
#Resuelve ecuaciones 
from scipy.integrate import odeint
```

```python
g=9.8
```

```python
#x01 solucion de la ecuacion (angulo inicial,velocidad inicial)
#l longitud cuerdad 
def mps(x01, L):
     
    def df1(f1,t):
        x1,v1=f1[0],f1[1]
        dx1=v1
        dv1=(-g/L)*x1
        return [dx1,dv1]
        
    t=np.arange(0,4,0.016)
    #soluci√≥n (funcion,condiciones(angulo,velocidad),tiempo)
    sol1=odeint(df1,x01,t)
    f1=sol1[:,0] 
    
    plt.grid()
    plt.plot(t,f1,color="b",lw=3,label="Lineal")
    
    return mps
```

```python
#(angulo,velocidad inicial),longitud
mps([3.2*np.pi/180,0],1)
```
